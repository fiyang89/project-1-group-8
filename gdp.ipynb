```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV file which was found on The World Bank https://data.worldbank.org/indicator/NY.GDP.MKTP.CD
file_path = "Resources/API_NY.GDP.MKTP.CD_DS2_en_csv_v2_4770391.csv"
gdp_df = pd.read_csv(file_path)

gdp_df
```

```python
# Reduce columns to Country Name and Years: 2010-2018
reduced_df = gdp_df[["Country Name","2010","2011","2012","2013","2014","2015","2016","2017","2018"]]
reduced_df
```

```python
# Restrict to 11 countries: Australia, Brazil, China, Germany, India, Israel, Japan, Russia = Russian Federation, South Africa, United Kingdom, United States
countries = [
    "Australia",
    "Brazil",
    "China",
    "Germany",
    "India",
    "Israel",
    "Japan",
    "Russian Federation",
    "South Africa",
    "United Kingdom",
    "United States"
]

df = reduced_df.loc[(reduced_df["Country Name"].isin(countries))]

df
```

```python
# rename column "Country Name" to "Country name" for later merge with population data frame, use .rename

column_df = df.rename(columns={"Country Name": "Country name"})
column_df
```

```python
# replace "Russian Federation" with "Russia", use .replace
rename_df = column_df.replace(["Russian Federation"], "Russia")

rename_df
```

```python
# In order to merge with population data frame, the year needs to be a value in a column, use .melt
df_unpivot = pd.melt(rename_df,
                     id_vars='Country name',
                     value_vars=[str(yr) for yr in range(2010,2019)],
                     var_name='Year', value_name='GDP')

# sort year column similar to population data frame, use .sort_values
sort_df = df_unpivot.sort_values(by=['Country name', 'Year'])
sort_df
```

```python
# line graph: x-axis: years 2010-2018, y-axis: GDP, separate color line for each country

for country in countries:
    df_gdp = sort_df[sort_df["Country name"] == country]
    plt.plot(df_gdp["Year"], df_gdp["GDP"], label=country)


# Place a legend on the chart in what matplotlib believes to be the "best" location
plt.legend(loc="best")

# Labels
plt.title("GDP of 11 Countries, 2010-2018")
plt.xlabel("Years")
plt.ylabel("GDP (US$)")

# Print our chart to the screen
plt.show()
```
