```python
#Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
#Csv Source - Read in File - https://ourworldindata.org/life-expectancy
df_life_expectancy = pd.read_csv("life-expectancy.csv")
```

```python
#Create Data Frame
df_life_expectancy
```

```python
#Rename Columns
df_life_expectancy = df_life_expectancy.rename(
    columns = {
    "Entity" : "Country name",
    "Life expectancy at birth (historical)" : "Life Expectancy at Birth"
}
)
df_life_expectancy
```

```python
#Set Countries needed and years needed
countries =[" United States", "United Kingdom", "China", "India", "South Africa", "Australia", "Brazil", "Japan", "Russia", "Israel", "Germany"]                         
df = df_life_expectancy.loc[(df_life_expectancy["Country name"].isin(countries)) &
                    (df_life_expectancy["Year"] > 2009) &
                    (df_life_expectancy["Year"] < 2019)]
df
```

```python
#Create Scatter Plot
plt.scatter(df["Country name"], df["Life Expectancy at Birth"])

#Set Title and Labels 
plt.title("Life Expectancy at Birth 2010-2018")
plt.ylabel("Life Expectancy at Birth")
plt.xlabel("Country")
plt.xticks(rotation=45)
plt.grid(True)

#Save Scatter Plot Locally
plt.savefig("Project Images/LifeExpectancyatBirth.png")
```
